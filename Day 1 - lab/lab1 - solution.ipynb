5) Check if a Number is Even or Odd
Ask the user for a number and determine whether it is even or odd.

```python
def checknum(n):
    if n % 2 == 0:
        print(f"{n} is even")
    else:
        print(f"{n} = is odd")

n = int(input("enter a number"))
checknum(n)
```

6) Find the Largest of Three Numbers
Take three numbers as input and print the largest one.

```python
def checkmax(a,b,c):
    max = a
    if b > c and b > max:
        max = b
    else:
        max = c
    return max

n1 = int(input("enter number 1"))
n2 = int(input("enter number 2"))
n3 = int(input("enter number 3"))
print(f"largest number is {checkmax(n1,n2,n3)}")

```

10) Prime Number Checker
Ask the user for a number and check if it's a prime number.

```python
def primenums (n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    else:
        return True
        
num = int(input("please enter a number"))
primenums(num)
```

11- Write a program that asks the user for:

 - The total number of students enrolled in a course

 - The desired number of students per group

The program should then calculate and display the total number of groups formed. If the students cannot be divided evenly, one group may contain fewer members than specified.

**Example output (1):**

- How many students on the course? 8
- Desired group size? 4
- Number of groups formed: 2


**Example output (2):**

- How many students on the course? 11

- Number of groups formed: 4


> Hint: The integer division operator // may be helpful in solving this problem.

```python
import math
n = int(input("n of students"))
pergroup = int(input("n per group"))
print(f"n of students is {n}")
print(f"desired group size {math.floor(n/pergroup)}")
print(f"n of groups formed {math.ceil(n/pergroup)}")

```
