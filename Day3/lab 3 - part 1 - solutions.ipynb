### **1) Calculator**

Please write a program which asks the user for two numbers and an operation. If the operation is add, multiply or subtract, the program should calculate and print out the result of the operation with the given numbers. If the user types in anything else, the program should print out nothing.

*Some examples of expected behaviour:*

Number 1: 10<br>
Number 2: 17<br>
Operation: add<br>

10 + 17 = 27

Number 1: 4<br>
Number 2: 6<br>
Operation: multiply<br>

4 * 6 = 24

```python
num1 = int(input("Number 1: "))
num2 = int(input("Number 2: "))
operation = input("Operation: ")

operations = {
    "add": ("+", num1 + num2),
    "subtract": ("-", num1 - num2),
    "multiply": ("*", num1 * num2),
    "divide": ("/", num1 / num2) 
}

if operation in operations:
    symbol, result = operations[operation]
    print(f"{num1} {symbol} {num2} = {result}")
```

### **2) Daily wages**

Please write a program which asks for the hourly wage, hours worked, and the day of the week. The program should then print out the daily wages, which equal hourly wage multiplied by hours worked, except on Sundays when the hourly wage is doubled.

Hourly wage: 8.5<br>
Hours worked: 3<br>
Day of the week: Monday<br>
Daily wages: 25.5 euros<br>

Hourly wage: 12.5<br>
Hours worked: 10<br>
Day of the week: Sunday<br>
Daily wages: 250.0 euros<br>

```python
def calc_daily_wages(hourly_wage, hours_worked, day):
    if day == "sunday":
        hourly_wage *= 2
    return hourly_wage * hours_worked

hourly_wage = float(input("enter hourly wage: "))
hours_worked = float(input("enter hours worked: "))
day = input("enter day of the week: ").lower()

daily_wages = calc_daily_wages(hourly_wage, hours_worked, day)

print(f"hourly_wage: {hourly_wage}")
print(f"hours_worked: {hours_worked}")
print(f"day: {day}")
print(f"daily_wages: {daily_wages} euros")
```

### **2) FizzBuzz**

Please write a program which asks the user for an integer number. If the number is divisible by three, the program should print out Fizz. If the number is divisible by five, the program should print out Buzz. If the number is divisible by both three and five, the program should print out FizzBuzz.

*Some examples of expected behaviour:*

Number: 9<br>
Fizz

Number: 20<br>
Buzz

Number: 7

Number: 45<br/>
FizzBuzz

```python
def divisible_check(number):
    if number % 3 == 0 and number % 5 == 0:
        return "FizzBuzz"
    elif number % 3 == 0:
        return "Fizz"
    elif number % 5 == 0:
        return "Buzz"

number = int(input("enter a number:"))
print(divisible_check(number))
```

### **3) PIN and number of attempts**

Please write a program which keeps asking the user for a PIN code until they type in the correct one, which is 4321. The program should then print out the number of times the user tried different codes.

PIN: 3245<br>
Wrong<br>
PIN: 1234<br>
Wrong<br>
PIN: 0000<br>
Wrong<br>
PIN: 4321<br>
Correct! It took you 4 attempts

**If the user gets it right on the first try,** the program should print out something a bit different:

PIN: 4321<br>
Correct! It only took you one single attempt!

```python
correct_pin = "4321"
attempts = 0
while True:
    pin = (input("enter pin"))
    attempts += 1
    print(f"PIN: {pin}")

    if pin == correct_pin:
        print("Correct!! bravoo")
        break
    else:
        print("Wrong")

print(f"it took you {attempts} attempts ")
```

6) Please write a function named greatest_number, which takes three 
arguments. The function returns the greatest in value of the three. 
[without using max function, and solve it in two ways] 


```python
def greatest_number(num1, num2, num3):
    if num1 > num2 and num1 > num3:
        return num1
    elif num2 > num1 and num2 > num3:
        return num2
    else:
        return num3
    
num1 = int(input("Enter first number: "))
num2 = int(input("enter second number: "))
num3 = int(input("enter third number: "))

print(f"the greatest number is {greatest_number(num1, num2, num3)}")
```

```python
def greatest_number(*args):
    greatest = args[0] 
    for num in args:
        if num > greatest:
            greatest = num
    return greatest

num1 = int(input("enter first number: "))
num2 = int(input("enter second number: "))
num3 = int(input("enter third number: "))

print("The greatest number is:", greatest_number(num1, num2, num3))
```

7) Write a function sum_series(n) that calculates the sum of the 
series: 
1 - 2 + 3 - 4 + 5 - ... up to n

```python
def sum_series(n):
    series = ""
    total = 0
    for i in range(1, n + 1):
        if i % 2 == 0:  
            total -= i
            series += f" - {i}"
        else:           
            total += i
            series += f" + {i}"
    return series, total

n = int(input("Enter n: "))
series_str, total = sum_series(n)
print("Sum of series:", total)
```

8) Write a Python function to calculate the factorial of a number (a 
non-negative integer). The function accepts the number as an 
argument.

```python
def factorial(n):
    if n < 0:
        return "put a positive numberrrr"
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

n = int(input("Enter a number: "))
print("Factorial:", factorial(n))
```
