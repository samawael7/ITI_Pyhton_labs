```python
squares = {i: i**2 for i in range(1, 6)}
for key, value in squares.items():
    print(f"{key}: {value}")

squares[6] = 36

for key, value in squares.items():
    print(f"{key}: {value}")

if 4 in squares:
    print("key 4 exists")
else:
    print("key 4 doesn't exist")
```

```python
students = {
    "sama": [80, 90, 85],
    "marwa": [80, 85, 80],
    "israa": [95, 90, 92]
}
averages = {}
for student, grades in students.items():
    avg = sum(grades) / len(grades)
    averages[student] = avg
    print(f"{student} : {round(avg, 1)}")

top_student = max(averages, key=averages.get)
print(f"highest average: {top_student}")
```

```python
cart = {}
cart["makeup"] = 560
cart["powder"] = 340
cart["perfume"] = 6000
for product, price in cart.items():
    print(f"{product}: {price:.2f}")
total = sum(cart.values())
print(f"total price: {total:.2f}")
```

```python
employees = [
    {"name": "sama", "age": 22, "department": "data e"},
    {"name": "ahmed", "age": 28, "department": "data e"},
    {"name": "mona", "age": 24, "department": "os"},
    {"name": "marwa", "age": 23, "department": "os"}
]
os_employees = []
for emp in employees:
    if emp["department"] == "os":
        os_employees.append(emp)
        
print("os employees:")
for emp in os_employees:
    print(emp["name"])

    
average_age = sum(emp["age"] for emp in employees) / len(employees)
print(f"average age of employees: {average_age:.2f}")
```
