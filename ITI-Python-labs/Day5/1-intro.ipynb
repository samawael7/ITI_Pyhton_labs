## 3. Object with default attribute values using Constructor
we **customize object creation** by defining an `__init__` **constructor** method. This constructor allocates memory for the object and initializes it with **default attribute values**: `name`, `email`, and `salary`.


> name, email, and salary called instance variable

```python
class Employee:
    """ Represents an Employee with default attributes: name, email, and salary."""

    def __init__(self):
        """ Initializes a new Employee object with default attribute values."""
        print("Object is being created")
        print(self)
        self.name = 'Ahmed Ali'
        self.email = 'ahmed@gmail.com'
        self.salary = 10000

emp3 = Employee()
emp4 = Employee()
print(emp3)

# emp4.email
```

- **`class` keyword**: Used to define a class.
- **Class Definition**: <span style="background:LemonChiffon">The name of the class followed by parentheses `()`</span> (if inheriting from another class). This is the **Class Header**.
- **Attributes**: Variables defined inside the class that represent properties of an object.
- **Methods**: Functions defined inside the class that represent behaviors of an object.
- **Constructor (`__init__`)**: A special method that initializes the object with specific values upon creation.
  - <span style="background:LemonChiffon">The constructor method is automatically called when an object is created, initializing the object with specific attributes</span>.
- **`self`**: A reference to the current instance of the class, allowing access to its attributes and methods.
- **Docstring (`""" """`)**: A special comment placed immediately after the class definition that describes the class and its purpose.
  - Enclosed in <span style="background:LemonChiffon">triple quotes</span>, allowing multi-line descriptions.
  - Python recognizes a **docstring** if placed immediately after the class definition.

### 5. Operations on Objects (Instances) using methods
- **An instance method like (`print_employee_info`)** is a function inside a class that can access and modify object attributes.


```python
class Employee:
    """Represents an Employee with name, email, and salary attributes."""

    def __init__(self, name, email, salary):
        """Initializes an Employee object with given attributes."""
        self.name = name
        self.email = email
        self.salary = salary


    def print_employee_info(self):
        """Instance method to print employee details."""
        print(f'Employee Name: {self.name} | Email: {self.email} | Salary: ${self.salary:,}')

    def calculate_salary(self):
        print(self.salary * .8)
```

```python
# Creating Employee objects
emp1 = Employee("ahmed ali", "ahmed@yahoo.com", 6000)
emp2 = Employee("Ali", "ali@yahoo.com", 10000)
emp1.calculate_salary()
emp2.calculate_salary()
```

```python
# Calling the instance method to print details
emp1.print_employee_info()
emp2.print_employee_info()
```

### 6. Counting Number of Objects in a Class

- We can track the number of objects created from a class using a **class variable.**
- **A class variable** belongs to the class itself, not to any specific object.


```python
class Employee:
    """Represents an Employee and tracks the number of objects created."""
    count = 0  # Class variable Shared among all instances to count instances Created

    def __init__(self, name, email, salary):
        """Initializes an Employee object and increments the count."""
        self.name = name
        self.email = email
        self.salary = salary        # print(self.__class__)
        self.__class__.count += 1   # Increment the count of Employee instances
        self.id = Employee.count

    def print_employee_info(self):
        """Displays the employee's details."""
        print(f'ID: {self.id} | Name: {self.name} | Email: {self.email} | Salary: ${self.salary:,}')
```

```python
# Creating Employee objects
emp1 = Employee("Ahmed", "ahmed@example.com", 5000)
```

```python
emp2 = Employee("Ali", "ali@example.com", 7000)
```

```python
# Calling the instance method
emp1.print_employee_info()
emp2.print_employee_info()
```

### 7. Dealing with Class methods and Factory Method

```python
class Employee:
    """Represents an Employee and tracks the number of instances created."""
    count = 0  # Shared class variable to track total employees

    def __init__(self, name="", email="", salary=0):
        """Initializes an Employee object with default or given values."""
        self.name = name
        self.email = email
        self.salary = salary
        self.__class__.count += 1  # Increment total count
        # Employee.count += 1
        self.id = Employee.count  # Assign unique ID

    def print_employee_info(self):
        """Displays the employee's details."""
        print(f'ID: {self.id} | Name: {self.name} | Email: {self.email} | Salary: ${self.salary:,}')

    @classmethod # Decorator --> changes function behavior to operate on the class level
    def display_employees_count(cls):
        """Displays the total number of employees."""
        print(f"Total number of employees: {cls.count}")

    @classmethod
    def create_emp(cls):
        """Creates a default employee instance."""
        return cls("Default User", "default@gmail.com", 30000)

    def create_employee(self):
        """Creates a new Employee instance with preset values."""
        return self.__class__("Ahmed", "ahmed@gmail.com", 20000)
```

> **Class method** call using class name

```python
emp = Employee("" , "" , 10000)
emp.create_employee()
```

```python
Employee.create_emp()
```

```python
Employee.display_employees_count()  # Display total employee count
```

```python
emp3 = Employee.create_emp()  # Creates a default employee instance
emp3.print_employee_info()
```

### 8. Static Method

```python
class Employee:
    def __init__(self, name, email, salary):
        self.name = name
        self.email = email
        self.salary = salary

    @staticmethod  # Static method doesn't depend on instance or class
    def cal_net_salary(salary):
        """Helper method to calculate net salary"""
        return salary * 0.8

    # def cal_net_salary(self):
    #     """Helper method to calculate net salary"""
    #     return self.salary * 0.8
```

```python
# Create an Employee instance
emp = Employee("test", "test@gmail.com", 10000)

# Use static method to calculate net salary
print(Employee.cal_net_salary(20000))  # Can be called directly using the class


print(Employee.cal_net_salary(emp.salary))  # Can also be called using the instance
```
