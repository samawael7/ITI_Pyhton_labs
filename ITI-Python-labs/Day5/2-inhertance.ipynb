```python
# Overriding Constructor and Using `super()`
class Person:
    def __init__(self, name):
        print("In parent class")
        self.name = name

    def display_name(self):
        print(f"Name: {self.name}")
#---------------------------------------------------------------------------
class Programmer(Person):
    def __init__(self, name, salary):
        print("In Child class")
        super().__init__(name)
        self.salary = salary

    def display_name(self):
        print(f"Name: {self.name} Salary: {self.salary}")
```

```python
programmer = Programmer()
```

```python
programmer = Programmer("Mona", 1000)  # True
programmer.display_name()
```

```python
# Basic Multiple Inheritance
class Person:
    pass

class Programmer:
    pass

class Engineer(Person, Programmer):
    pass
```

```python
# Constructor in One Parent Class
class Person:
    pass

class Programmer:
    def __init__(self):
        print("Programmer object is being created")

class Engineer(Person, Programmer):
    pass
```

```python
engineer = Engineer()  # Will not call Programmer’s constructor because Engineer doesn't define its own
```

```python
# Both Parent Classes Have Constructors
class Person:
    def __init__(self , name):
        print("Person object is being created")
        self.name = name


class Programmer:
    def __init__(self, email):
        print("Programmer object is being created")
        self.email = email

class Engineer(Programmer, Person):
    def __init__(self, name, email, age):
        print("Engineer object is being created")
        super.__init__(email)
        # Person.__init__(self, name)
        # Programmer.__init__(self, email)
        self.age = age
```

```python
   # Will call only A’s constructor due to method resolution order (MRO)
   engineer = Engineer("Mona", "M@gmail", 20)
```
