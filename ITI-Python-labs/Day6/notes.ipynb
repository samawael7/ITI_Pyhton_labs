OOP (Encapsulate)

```python
class Employee:
    def __init__(self, name, email, salary):
        self.name = name #  public attribute
        self._email = email # protected attribute
        self.__salary = salary # private attribute
    
    @property #بتخلينا نستخدمها كأنها attribute مش function
    def salary(self):
        # getter for private attribute to retrieve salary
        return self.__salary
    
    @salary.setter
    def salary(self, sal):
        # setter example validation: salary must be positive
        if isinstance(sal, (int, float)) and sal > 0:
            self.__salary = sal
        else:
            raise ValueError("Salary must be a positive number")
        
    def get_email(self):
        # getter for protected attribute to retrieve email
        return self._email
        

# Example usage
emp = Employee("Ahmed", "ahmed@example.com", 50000)

print(emp.salary)
emp.salary = 60000
print(emp.salary)

print(emp.get_email())
```

Polymorphism

```python
# polymorphism + method overriding example

# parent class
class Person:
    def __init__(self, name):
        self.name = name  # public attribute for the person's name

    def print_info(self):
        # method to print basic info about the person
        print(f"my name is {self.name}")  # prints the name


# child class (inherits from Person)
class Engineer(Person):
    def __init__(self, name, email):
        super().__init__(name)  # call parent constructor to set name
        self.email = email      # additional attribute for Engineer

    def print_info(self):
        # method overriding: same method name as in parent, but different implementation
        super().print_info()  # optional: call parent version to reuse code
        print(f"my email is {self.email}")  # additional info specific to Engineer


# creating objects
p = Person("Ahmed")                # create a Person object
e = Engineer("Sara", "sara@mail.com")  # create an Engineer object

# calling methods
p.print_info()  # calls Person.print_info
print("-----")
e.print_info()  # calls Engineer.print_info
# here polymorphism happens: same method name (print_info) behaves differently
```
