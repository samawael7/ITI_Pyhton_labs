### 2. Local Variable Scope

```python
def print_course_name():
    course = "Python"
    print(f"Course: {course}")

print(course)
```

> `course` is `Local` variable, accessible `only inside` the function

```python
print_course_name()
```

```python
print(course)  #
```

> This would raise an error as 'course' is `local variable` not defined `globally`

```python
print_course_name()
```

### 4. Modifying a Global Variable Inside a Function

```python
course_name = 'Python' # Global variable

def modify_course_name():
    course_name = 'Django'  # create new local variable Modify the global variable
    print(f"Modified 'course_name' inside function: {course_name}")
```

```python
modify_course_name()
```

```python
print(course_name)
```

> a **global variable** is accessible from anywhere in the script. However, if you want to **modify** a global variable **inside a function**, you must use the `global` keyword to modify value of global variable `not` created `new local variable`.

```python
course_name = 'Python' # Global variable

def modify_course_name():
    global course_name
    course_name = 'Django'
    print(f"Modified 'course_name' inside function: {course_name}")
```

```python
modify_course_name()
```

```python
print(course_name)
```

### 5. Nested Functions and Scope

- A **nested function** is a function defined inside another function.
- The **inner function** can access variables from the **outer function** due to Pythonâ€™s `lexical scoping` (closure).

```python
def display_salary_details():
    salary = 50000  # Local variable in outer function

    def show_inner_salary():
        # Access local variable from inside the inner function
        print(f"Salary Value In Inner Function : {salary}")

    show_inner_salary()
    print(f"Salary Value In Outer Function : {salary}")
```

```python
display_salary_details()
```

---

### 6. Nonlocal Keyword Usage

```python
def outer_function_with_nonlocal():
    salary = 50000

    def modify_salary():
        nonlocal salary  # Use the parent function's local variable
        salary = 100000  # Modify the local variable
        print(f"Modified salary inside inner function: {salary}")

    modify_salary()
    print(f"Final salary after modification: {salary}")
```

```python
outer_function_with_nonlocal()
```

> Modifications to salary inside `show_inner_salary` without using `nonlocal` will create a `new local variable` instead of `modifying the outer one`.

### 7. Using And Modify Global Variable  Inside a Nested Function

```python
global_message = "Welcome"  # Global variable

def outer_function_with_global():
    def modify_global():
        global global_message  # Declare that we are modifying the global variable
        global_message = "Hello, Python!"  # Modify global variable
        print(f"Modified inside nested function: {global_message}")

    modify_global()
```

```python
outer_function_with_global()
```

```python
print(f"Global variable after modification: {global_message}")
```
