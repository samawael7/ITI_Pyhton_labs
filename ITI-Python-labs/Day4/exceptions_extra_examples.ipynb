### **This file is extra example on handling exceptions topic**

```python
# Example 1: Basic Try-Except
try:
    print(name)  # 'name' variable should be defined
except NameError:
    print("NameError: Variable 'name' is not defined")
else:
    print("No error occurred")
finally:
    print("This block always executes")
```

```python
# Example 2: Handling ZeroDivisionError
try:
    num = 10
    print(num / 0)  # Division by zero error
except ZeroDivisionError:
    print("ZeroDivisionError: Cannot divide by zero")
finally:
    print("Execution completed\n")

print("========================================")

# Example 3: Catching General Exception with Error Message
try:
    print(num / 0)
except Exception as e:
    print(f"Exception Occurred: {e}")
finally:
    print("Execution completed\n")

print("========================================")

# Example 4: Multiple Except Blocks
try:
    num = 10
    result = num / 0  # Division by zero
except NameError as ne:
    print(f"NameError: {ne}")
except ZeroDivisionError as ze:
    print(f"ZeroDivisionError: {ze}")
except Exception as e:
    print(f"Other Exception: {e}")
finally:
    print("Execution completed\n")

print("========================================")

# Example 5: Using Else Block
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("ValueError: Invalid input, please enter a number")
else:
    print("No errors occurred")
    num *= 10
    print(f"Result: {num}")
finally:
    print("Process finished\n")

print("========================================")

# Example 6: Finally Block Execution
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("ValueError: Invalid input")
else:
    print("No errors, processing the number...")
    num *= 10
    print(f"Processed Number: {num}")
finally:
    print("The process is completed (finally block always executes)")

print("===== Bye Bye =====\n")

print("========================================")

# Example 7: Function Handling Exceptions
def ask_for_int():
    """Function to get an integer input safely."""
    try:
        num = int(input("Please enter an integer: "))
    except ValueError:
        print("ValueError: Invalid input, please enter an integer")
        return False
    else:
        return num

# print(ask_for_int())

print("========================================")

# Example 8: Function with Finally Block
def ask_for_int_finally():
    """Function demonstrating the finally block in return scenarios."""
    try:
        num = int(input("Please enter an integer: "))
    except ValueError:
        print("ValueError: Invalid input")
        return False
    else:
        print("Returning value from function")
        return num
    finally:
        print("Finally block executes before return")

# print(ask_for_int_finally())

print("========================================")

# Example 9: Finally Block Overrides Return Value
def ask_for_int_override():
    """Function demonstrating how finally block can override return values."""
    try:
        num = int(input("Please enter an integer: "))
    except ValueError:
        print("ValueError: Invalid input")
        return False
    else:
        print("Returning value from function")
        return num
    finally:
        print("Finally block executed, overriding return value")
        return "bye"

print(ask_for_int_override())
```
