> ### 6. <span style="background:LavenderBlush">ZeroDivisionError : Division by zero</span>


```python
result = 10 / 0
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Division by zero!")
except Exception as e :
    print(e)
```

```python
try:
    num1 = int(input("Enter a number: ")) # ValueError
    num2 = int(input("Enter another number: ")) # ValueError
    result = num1/num2 # ZeroDivisionError
    print(result)
except ZeroDivisionError:
    print("division by zero, please enter a valid number > 0.")
except ValueError:
    print("Invalid input! Please enter a valid number.")
except Exception as e : # else
    print(e)
```

```python
try:
    x = 10 / 0
except (ZeroDivisionError, ValueError, TypeError):
    print("An error occurred, division by zero!")
except NameError:
    print("Name error")
except Exception as e : # else
    print(e)
```

```python
try:                    # Attempt to execute a block of code.
    pass
except Exception as e:  # Catch and handle specific exceptions.
    pass
else:                   # Executes if no exception occurs.
    pass
finally:                # Always executes, regardless of whether an exception occurred.
    pass
```

### Using else for Code That Runs If No Exception Occurs

```python
num = int(input("Enter numeric value"))  # No error
```

```python
try:
    num = int(input("Enter numeric value"))  # No error 10
except ValueError:
    print("Invalid number!")
else:
    print(f"Valid number: {num}") # Runs else block if no error happens
finally:
    print("Goodbye!")
```

```python
try:
    num = int(input("Enter numeric value"))  # No error 10k
except ValueError:
    print("Invalid number!")
else:
    print(f"Valid number: {num}") # Runs else block if no error happens
finally:
    print("Goodbye!")
```

```python
def sum_numbers(num1:int, num2:int):
    """
    Adds two numbers only if both are integers.

    :param num1: First number
    :param num2: Second number
    :return: Sum of num1 and num2
    :raises TypeError: If inputs are not integers
    """
    print(f"num1 = {num1}, num2 = {num2}")

    if not isinstance(num1, int) or not isinstance(num2, int):
        # print("Both num1 and num2 must be integers")
        raise TypeError("Both num1 and num2 must be integers")

    return num1 + num2
```

```python
sum_numbers(10, "20")
```

```python
try:
    print(sum_numbers("10", "20"))
except TypeError as e:
    print(f"Error: {e}")

print("=" * 50)
```

```python
def divide_numbers(num1, num2):
    """
    Divides num1 by num2, handling division by zero.
    """
    if num2 == 0:
        raise ZeroDivisionError("(num2) cannot be zero.")

    print(f"num1 = {num1}, num2 = {num2}")
    print(f"Result: {num1 / num2}")
    return num1 / num2
```

```python
divide_numbers(4, 0)
```

```python
# Example of handling ZeroDivisionError
try:
    divide_numbers(4, 0)
except ZeroDivisionError as e:
    print(f"Error: {e}")

print("=" * 50)
```

```python
# Example with valid inputs
try:
    print(f"Valid Division: {divide_numbers(10, 2)}")
except ZeroDivisionError as e:
    print(f"Error: {e}")
```
