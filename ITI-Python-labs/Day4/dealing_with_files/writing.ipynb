## Writing Data to a File in Python ✍️

### Opening a File in Write Mode (`'w'`)
- If the file **does not exist**, Python **creates** a new file.
- If the file **exists**, it **overwrites** the existing content.

---
### Methods for Writing to a File
---

#### 1. Writing a Single String (`write()`)
    Writes a single string to the file.

```python
from os import write

try:
    file = open("cv.txt", "w")  # Open file in write mode
except Exception as e:
    print(f"Error: {e}")
else:
    file.write("Hello, This Is My CV!\nHello Python")  # Write a string to the file
    print("Data written successfully.")
finally:
    file.close()  # Close the file
```

> <span style="background:LemonChiffon">Can Write content to the file in else block also</span>

```python
try:
    file = open("cv.txt", "w")
except Exception as e:
    print(f"Error: {e}")
else:
    file.write("New Version Of My Cv :\n")
    print("Data written successfully.")
finally:
    file.close()
```

---

#### 2. Writing Multiple Lines (writelines())
    Writes a list of strings to the file.

```python
try:
    file = open("cv.txt", "w")
except Exception as e:
    print(f"Error: {e}")
else:
    lines = [
        "Name: Mona\n",
        "Age: 25\n",
        "Faculty: Engineering\n",
    ]
    file.writelines(lines)  # Write multiple lines
    print("Multiple lines written successfully.")
finally:
    file.close()
```
