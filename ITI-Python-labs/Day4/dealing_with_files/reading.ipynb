## Reading Data from a File in Python

### Opening a File in Read Mode (`'r'`)
- The file **must exist**; otherwise, a `FileNotFoundError` occurs.
- Reads content as a **single string** or **line-by-line**.

### Steps for Reading a File
1. Open the file.
2. Read the content.
3. Close the file.


---
### Methods for Reading a File
---

```python
# 1- open file
file = open("users.txt", "r")

# 2- read string
print(file.read())
print(type(file.read()))

# 3- close file
file.close()
```

#### 1. Read Entire Content (`read()`)
    Reads the entire file as a single string.

```python
file = open("users.txt", "r")  # Open file in read mode
content = file.read()  # Read file content as a string
print(content)
file.close()  # Close the file
```

```python
print(type(content))
print(type(file))
```

#### ðŸ“ **Note**
> <span style="background:LemonChiffon">Always handle file-related errors Like `FileNotFoundError` to prevent program crashes.</span>

```python
try:
    file = open("users.txt", "r")
except FileNotFoundError:
    print("Error: File not found. Please check the filename.")
else:
    print(file.read())
finally:
    file.close()
```

---

#### 2. Read as a List of Lines (readlines())
    Reads all lines into a list.

```python
try:
    file = open("users.txt", "r")
except FileNotFoundError:
    print("Error: File not found. Please check the filename.")
else:
    lines = file.readlines()  # Returns a list of lines # Check the type of lines
    print(lines)
finally:
    file.close()
```

> Direct conversion file object to a list

```python
try:
    file = open("users.txt", "r")
except FileNotFoundError:
    print("Error: File not found. Please check the filename.")
else:
    data_list = list(file)  # without using readlines
    print(data_list)
finally:
    file.close()
```

---

#### 3. Read Line-by-Line (for loop)
    Iterates over the file to read line-by-line.

```python
try:
    file = open("users.txt", "r")
except FileNotFoundError:
    print("Error: File not found. Please check the filename.")
else:
    for line in file:
        print(line.strip())
finally:
    file.close()
```

---

#### 4. Using seek(0) to Reset File Pointer
    If you need to read again from the beginning:

```python
file = open("users.txt", "r")
print(file.read())  # Reads entry file and pointer in last line if you read again nothing show
file.seek(0)  # Reset pointer to beginning
print(file.read())  # Reads again
file.close()
```

---
#### âœ… Best Practices for Error Handling

```python
try:
    file = open("users.txt", "r")
except FileNotFoundError:
    print("Error: File not found. Please check the filename.")
except PermissionError:
    print("Error: Insufficient permissions to access the file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else:
    print(file.read())
finally:
    file.close()
```
