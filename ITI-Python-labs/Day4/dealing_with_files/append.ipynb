## Appending Data to a File in Python

### Opening a File in Append Mode (`'a'`)

- If the file **does not exist**, Python **creates** a new file.
- If the file **exists**, new content is **added at the end** **without overwriting** existing content.

```python
file = open("cv.txt" , "a")
file.write("Hi\n")
file.close()

with open("cv.txt" , "a") as file:
    file.write("Hi\n")
```

```python
with open("cv.txt", "a") as file: # close file automatically
    lines = [
        "Name: Mona\n",
        "Age: 25\n",
        "Faculty: Engineering\n",
    ]
    file.writelines(lines)  # Write multiple lines

print("Multiple lines written successfully.")
```

```python
file_name = "cv.txt"  # File path

# Open file in append mode and write content and automatic close it
with open(file_name, "a") as file:
    print(f"File '{file_name}' opened successfully for appending data.")

    file.write("------------------------------------------------\n")
    file.write("Courses :\n")
    file.writelines([
        "Php\n",
        "Python\n",
        "Html\n"
    ])

print(f"Data successfully appended to '{file_name}'.")
```

```python
file_name = "cv.txt"  # File path

try:
    file = open(file_name, "a")
    file.write("------------------------------------------------\n")
    file.write("Certificates\n")
    file.writelines([
        "Aws\n",
        "Maharatech\n"
    ])
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else:
    print(f"Data successfully appended to '{file_name}'.")
finally:
    file.close()
    print("File closed successfully.")
```
