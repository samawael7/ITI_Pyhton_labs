```python
import json

class queue:
    def __init__(self):
        self.items = []

    def insert(self, value):
        self.items.append(value)

    def pop(self):
        if self.is_empty():
            print("warning: queue is empty!")
            return None
        return self.items.pop(0)

    def is_empty(self):
        if len(self.items) == 0:
            return True
        else:
            return False

class queueoutofrangeexception(Exception):
    pass

class namequeue(queue):
    queues = {}

    def __init__(self, name, size):
        super().__init__()
        self.name = name
        self.size = size
        namequeue.queues[name] = self

    def insert(self, value):
        if len(self.items) >= self.size:
            raise queueoutofrangeexception("queue '{}' is full!".format(self.name))
        super().insert(value)

    @classmethod
    def search_queue(cls, name):
        return cls.queues.get(name)

    @classmethod
    def save(cls, filename):
        data = {}
        for name, q in cls.queues.items():
            data[name] = {"size": q.size, "items": q.items}
        with open(filename, "w") as f:
            json.dump(data, f, indent=4)

    @classmethod
    def load(cls, filename):
        cls.queues.clear()
        with open(filename, "r") as f:
            data = json.load(f)
        for name, info in data.items():
            q = cls(name, info["size"])
            q.items = info["items"]

q1 = namequeue("numbers", 3)
q1.insert(10)
q1.insert(20)
q1.insert(30)

q2 = namequeue("letters", 2)
q2.insert("a")
q2.insert("b")

print(q1.pop())
print(q1.pop())
print(q1.pop())
print(q1.pop())

found_q = namequeue.search_queue("letters")
print(found_q.items)

namequeue.save("queues.json")
namequeue.load("queues.json")
loaded_q = namequeue.search_queue("numbers")
print(loaded_q.items)
```
