### **Loops**
  * The `for` loop is `used to iterate over a sequence ` (such as a list, tuple, string, or range) or `any iterable` object.
  * It allows you to `execute a block of code repeatedly for each` item in the sequence.
![Loops](images/working-continue-statement-python.png "Loops")

````python
# 1 - 10



for value in range(0,11):
    if value % 2 == 0:
        print(value)


string = "ITI"

for char in string:
    print(char)


````

2. while loop
  * The `while` loop allows you to repeatedly `execute a block of code` as long as a `given condition remains true`.
  * It is useful when you want to repeat a certain task until a specific condition is no longer met.
**The structure of a while loop is as follows**

````python
num = 0
while num <= 10:
    print(num)
    num += 1
````

````python
# check password
password = ""
while password != "1234":
    password = input("Enter your password")

print("password correct")
````
