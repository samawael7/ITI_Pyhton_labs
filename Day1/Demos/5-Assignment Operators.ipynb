### Assignment operators

### **(=) assign**


```python
x = 20
print(x)
```

### **+= add and assign**

```python
x = 20
x = x + 2


x += 2
print(x)
```

### **-= subtract and assign**

```python
x = 20
# x = x - 2
x -= 2
print(x)
```

### *= **multiply and assign**

```python
x = 20
x *=  2
print(x)
```

### **/= divide and assign**

```python
x = 20
x = x / 2
# x /= 2
print(x)
```


### **%= get modulus and assign**

```python
x = 20
x %= 2 # x = x % 2
print(x)
```


### **//= floor divide and assign**

```python
x = 20
x //= 2 # x = x // 2
print(x)
```

### **= **get exponent and assign**

```python
x = 3
x = x ** 2
print(x)
```
