### String Case Manipulation Methods

Python provides built-in methods to change the case of a string:

- **`lower()`** → Converts all characters to lowercase.
- **`upper()`** → Converts all characters to uppercase.
- **`capitalize()`** → Converts the first character to uppercase and the rest to lowercase.
- **`title()`** → Capitalizes the first letter of each word in the string.

### Example:


````python
name = input("Enter your name:")
````

````python
print(name)
````

````python
name.title()
````

````python
text = "Hello world, welcome to Python!"

# Apply string case methods
print("Original Text: ", text)
print("Lowercase: ", text.lower()) # return new string
print("Uppercase: ", text.upper())
print("Capitalized: ", text.capitalize())
print("Title Case: ", text.title())
````

### **String Content Examination Methods**

Python provides built-in methods to check the type of characters in a string:

1. **`isalpha()`** → Returns `True` if the string consists only of alphabetic characters (`a-z` or `A-Z`), without spaces or digits.
2. **`isupper()`** → Returns `True` if all characters are uppercase letters.
3. **`islower()`** → Returns `True` if all characters are lowercase letters.
4. **`isspace()`** → Returns `True` if the string consists only of whitespace characters (spaces, tabs, or newlines).

#### **Example:**


````python
num1 = input("Please enter a string: ")
num2 = input("Please enter another string: ")

# Check for alphabetic characters
print("is alpha:", num1.isalpha())  # True

# Check for uppercase
print("is upper:", num1.isupper())  # True

# Check for lowercase
print("is lower:", num2.islower())  # True

# Check if the string consists only of spaces
name = "                               "
print("is space:", name.isspace())  # True
````

### **Nubmers -> Finding Minimum, Maximum, and Rounding Numbers**

Python provides built-in functions to perform common numeric operations:

1. **`min()`** → Returns the smallest value in a list of numbers.
2. **`max()`** → Returns the largest value in a list of numbers.
3. **`round()`** → Rounds a floating-point number to the nearest integer or a specified number of decimal places.

#### **Syntax:**
```python
min(value1, value2, ...)
max(value1, value2, ...)
round(number, digits)  # 'digits' is optional; default is 0 decimal places


````python
numbers = [3, 7, 2, 8.5, 10, 4]

# Finding min and max
print("Minimum number:", min(numbers))  # Output: 2
print("Maximum number:", max(numbers))  # Output: 10
````

````python
# Rounding numbers
num = 3.4896666666
print("Rounded to nearest integer:", round(num))    # Output: 4
print("Rounded to 2 decimal places:", round(num, 2))  # Output: 3.68
````

````python
a,b,c,d = 10, 2, 8,5
````

````python
# Using min and max with individual values
print("Min of", min(a,b,c,d))
````

````python
print("Max of", max(a,b,c,d))
````
