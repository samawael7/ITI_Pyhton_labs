access a char in a string at a specific index
string = immuatable data type , you cannot change char you can only reassign the whole string

```python

name = "sama wael abdou"
print(name[0:4]) # from index 0 to 3
print(name[5:9]) # from index 5 to 8
print(name[10:15]) # from index 10 to 14
print(name[0:]) # full string
print(name[:]) # full string
print(name[:1]) # first character
print(name[1:]) # from index 1 to end
print(name[-4:]) # last 4 characters
print(name[-15:-11]) # from start to index -11
print(name[-15:4]) # from start to index 4
print(name[::2]) # every second character , start from 0 skip 1
print(name[1::2]) # every second character starting from index 1
print(name[::-1]) # reverse string
print(name[-1]) # last character
print(name[::-2]) # reverse string with step 2
print(name[::3]) # every third character
print(name[1::3]) # every third character starting from index 1
```

using COUNT() to find occurrences of a substring

```python
text = "python, python, python"
#text.count("o") 
print(text.count("o")) # count occurrences of 'o'
print(text.count(",")) # count occurrences of ','
print(text.count("python")) # count occurrences of 'python'
print(text.index(",")) # index of first occurrence of ','
```

String manipulation

```python
lower() # convert to lowercase
upper() # convert to uppercase
strip() # remove leading and trailing whitespace
capitalize() # capitalize first character
title() # capitalize first character of each word
replace()  # replace substring with another substring
split() # split string into list based on delimiter
find() # find substring and return index
join() # join list of strings into a single string with delimiter
format() # format string with placeholders
startswith() # check if string starts with a substring
endswith() # check if string ends with a substring
len() # get length of string
```

```python
name = "    sama wael   "
name = name.strip()
print(name)
print(name.upper()) # SAMA
print(name.lower()) # sama
print(name.capitalize()) # Sama
print(name.title()) # Sama
print(name.replace("a", "o")) # somo
print(name.split(" ")) # ['sama']
```

String Concatination


```python
    # 1. using + operator 
    # 2. using f-string  // most used (insert var directly into a string)
    # 3. using join()

str1 = "hello"
str2 = "sama"
str3 = str1 + " " + str2
print(str3) # hello sama


str4 = f"{str1} {str2}" 
print(str4) # hello sama

name = "sama"
age = 25
print(f"your name is {name} age is {age}") # f-string 

joined_str = " ".join([str1, str2])
print(joined_str) # hello sama

# looping through a string
for char in name:
    print(char)
# output each character in a new line
```

```python
name = "sama wael"
for index in range(len(name)):
    print(f"index {index} : {name[index]}")
# output index and character at that index

i = 0 
while i < len(name):
    print(f"index {i} : {name[i]}")
    i += 1
```

Lists

```python
list = ["red", "blue", "green"]
print(list)

print(list[0]) # red

print(list.index("green")) 

print("wael" in list) # True


list.append("new item") # add item to end of list
print(list)

print(len(list)) # length of list

print(list.count("red")) # count occurrences of 'red'

list.remove("blue")
print(list)

list.pop() # removes last item
print(list)

list.insert(1, "inserted item") # insert at index 1
print(list)

list.sort() # sort list
print(list)

list.reverse() # reverse list
print(list)

print(list.clear()) # clear list
print(list)

print(type(list)) # <class 'list'>

name = ["sama", "wael"]
print(len(name)) # length of string
print(len(list)) # length of list
print(min(name)) # minimum character
print(max(name)) # maximum character
print(sorted(name)) # sorted characters
print(sorted(list)) # sorted items
print(sum([1, 2, 3, 4, 5])) # sum of numbers in list
print(sum([10, 20, 30])) # sum of numbers in list
 # duplicate list
print(name * 3) # duplicate string
```
