3- Write a program that remove all vowels from the input word and
generate a brief version of it

---

```python
word = input("Enter a word: ")
vowels = "iouae"

result = ""
for char in word.lower():
    if char not in vowels:
        result+=char
print("Brief version:", result)
```

7-  Write a Python script to:
- Take a string input from the user.
- Convert the string to uppercase.
- Count the number of vowels in the string.
- Reverse the string.
- Check if the string is a **palindrome**
--  example: madam

```python
text = input("enter a string: ")
text = text.upper()
vowels = "AEIOU"
vcount = 0

for char in text:
    if char in vowels:
        vcount += 1
reversed_text = text[::-1]
is_palindrome = text == reversed_text

print("uppercase:", text)
print("vowel count:", vcount)
print("reversed:", reversed_text)
print("palindrome:", is_palindrome)

```

# Sets in python

### 1- Create a set of the first five even numbers. Write a script to:

- Add a new number to the set.
- Remove a number from the set.
- Check if a specific number exists in the set.
- Perform a union and intersection with another set `{2, 4, 6, 8, 10}`.


```python
set1 = {2, 4, 6, 8, 10}

set1.add(12)
set1.remove(4)

print(set1)

if 6 in set1:
    print("6 exists")
else:
    print("6 not found")

set2 = {2, 4, 6, 8, 10}

print(set1.union(set2))
print(set1.intersection(set2))
```

### 2- Unique Visitors Tracker
Create a set to track unique website visitors. Write a script to:

- Add new visitors to the set.
- Check if a visitor has already visited.
- Remove a visitor from the set.


```python
visitors = set()

visitors.add("sama")
visitors.add("ahmed")

name = input("enter visitor name: ")
if name in visitors:
    print("exsisting visitor", name)
else:
    visitors.add(name)

visitors.remove("ahmed")

print("current visitors:", visitors)
```

### 3- Set-Based Password Security Checker

## Task:
Implement a function that checks if a password is weak by comparing it to a predefined set of common passwords.

**Common passwords:**
- `"123456"`, `"password"`, `"123456789"`, `"12345"`, `"12345678"`, `"qwerty"`
- `"1234567"`, `"111111"`, `"123123"`, `"abc123"`, `"password1"`, `"1234"`


```python
weak_passwords = {
    "123456", "password", "123456789", "12345",
    "12345678", "qwerty", "1234567", "111111",
    "123123", "abc123", "password1", "1234"
}

your_pass = input("Enter password: ")

if your_pass in weak_passwords:
    print("Weak password")
else:
    print("Strong password")
```

## **Task 2: Tuple-Based Stock Price Analysis**
Given a tuple of stock prices for a week, write a script to:

- Find the highest price.
- Find the lowest price.
- Calculate the average price.
    -  `Output`:
    -	Highest Price: 162
    -	Lowest Price: 145
    -	Average Price: 154.00


```python
fruits = ("apple", "banana", "orange", "mango", "watermelon")

print("first fruit:", fruits[0])
print("last fruit:", fruits[-1])
print("first three fruits:", fruits[:3])
print("count of apple:", fruits.count("apple"))

try:
    fruits[0] = "guava"
except:
    print("tuples are immutable can't be changed.")



stock_prices = (150, 155, 162, 148, 145, 158, 160)
print("highest Price:", max(stock_prices))
print("lowest Price:", min(stock_prices))
average = sum(stock_prices) / len(stock_prices)
print("average Price: {:.2f}".format(average))
```
