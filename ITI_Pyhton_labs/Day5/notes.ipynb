```python
class Employee:

    count = 0  # class variable to keep track of number of employees

    def __init__(self, name, email, salary): 
        # Constructor to initialize employee attributes
    
        self.name = name
        self.email = email
        self.salary = salary
        self.__class__.count += 1  # class variable access directly by class name
    
    def print_info(self):
        print(f"Name: {self.name}, Email: {self.email}, Salary: {self.salary}")

    
    def calculate_bonus(self):
        bonus = self.salary * 0.10
        return bonus
    
    #class method instead of instance method
    @classmethod # << decorator >> to define a class method called
    def count_std(cls): # cls refers to the class itself
        print(cls.count)


    @staticmethod # static method decorator doesn't depend on class or instance
    def calc_salary(salary, bonus): #no self or cls parameter
        return salary + bonus   

        
        
emp1 = Employee("sama", "sama@gmail.com", 50000)
emp1.print_info()
emp1.calculate_bonus()
print(emp1.count) 
print(Employee.calc_salary(emp1.salary, emp1.calculate_bonus()))

emp2 = Employee("ali", "ali@gmail.com", 60000)
print(emp2.__dict__)
print(emp2.count)
Employee.count_std()
```

Inheritance

```python
class Person:

    count = 0  # class variable to keep track of number of persons

    def __init__(self, fname, lname): 
        # Constructor to initialize person attributes
    
        self.fname = fname
        self.lname = lname
        self.__class__.count += 1  # class variable access directly by class name

    def display_fullname(self):
        print(f"Full Name: {self.fname} {self.lname}")


class programmer(Person): # inheritance from Person class
    def __init__(self, fname, lname, salary):
        super().__init__(fname, lname)  # call the constructor of the parent class كأني كتبت person(fname, lname)
         # initialize programmer-specific attribute
        self.salary = salary


programmer1 = programmer("sama", "wael", 5000)
print(programmer1.fname)
programmer1.display_fullname()
programmer1.__dict__

    
```

Multiple inheritance

```python
class a:
    pass

class b:
    def __init__(self, name):
        self.name = name
        print("class b init")

class c(a, b): # multiple inheritance
    def __init__(self, name, age):
        super().__init__(name)  # call constructor of the first parent that has constructor
        self.age = age
        print("class c init")



obj = c("sama", 23)

print(obj.name)
print(obj.age)
```
