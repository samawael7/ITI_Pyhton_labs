### **Falsy values**

> - #### Falsy values are values that evaluate to False in a boolean context.
> - #### Falsy values include empty sequences (lists `[]`, tuples `()`, strings `""`, dictionaries `{}`, sets `set()`), zero `0` in every numeric type, `None`, and `False`.

#### Examples of Falsy Values:
- `0`
- `False`
- `None`
- `[]` (empty list)
- `()` (empty tuple)
- `""` (empty string)
- `{}` (empty dictionary)
